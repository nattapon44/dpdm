# **2) สำรวจค่าสูญหาย**

- ค่าสูญหายถูกแทนด้วยสัญลักษณ์อะไร?

```python
# "?"
```

- มีค่าสูญหายจำนวนเท่าไหร่ในแต่ละคอลัมน์

```python
adult.isna().sum()
```

- ในกรณีนี้ให้แทนค่าสูญหายด้วย mean/median/mode

```python
adult[['workclass']] = adult[['workclass']].fillna(adult[['workclass']].mode().iloc[0,0])
adult[['occupation']] = adult[['occupation']].fillna(adult[['occupation']].mode().iloc[0,0])
adult[['native-country']] = adult[['native-country']].fillna(adult[['native-country']].mode().iloc[0,0])
```

```python
adult.isna().sum()
```

- อาชีพใด (occupation) มีจำนวนตัวอย่างที่มีรายได้ >50K มากที่สุด

```python
adult[adult['income'] == '>50K'].groupby('occupation').size()
```

```python
high_income_occupation = adult[adult['income'] == '>50K'].groupby('occupation').size()
max_high_income_occupation = high_income_occupation.idxmax()

print("อาชีพที่มีจำนวนตัวอย่างที่มีรายได้ >50K มากที่สุด:", max_high_income_occupation)
```
